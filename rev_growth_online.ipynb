```python
import matplotlib.pyplot as plt
import numpy as np

# Year-over-Year Revenue Growth
years = np.array([2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
revenues = np.array([18, 20, 22, 25, 28, 30, 31, 33])

plt.figure(figsize=(10, 5))
plt.plot(years, revenues, marker='o')
plt.title('Year-over-Year Revenue Growth of Online Food Delivery')
plt.xlabel('Year')
plt.ylabel('Revenue (in billions USD)')
plt.grid(True)
plt.show()

# Funding and Market Position
companies = ['Rappi', 'Delivery Hero', 'iFood']
funding = [2.2, 1.5, 1.0]

plt.figure(figsize=(8, 5))
plt.bar(companies, funding, color=['blue', 'green', 'red'])
plt.title('Total Funding Received by Key Players')
plt.xlabel('Company')
plt.ylabel('Funding (in billions USD)')
plt.show()
```
