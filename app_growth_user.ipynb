```python
import matplotlib.pyplot as plt

# Years
years = [2018, 2019, 2020, 2021, 2022, 2023]

# Data for each app
user_base_pedidos_ya = [50, 70, 120, 180, 240, 300]
user_base_hugo = [30, 50, 80, 130, 200, 250]
user_base_uber_eats = [40, 60, 100, 160, 220, 280]

plt.figure(figsize=(10, 5))
plt.plot(years, user_base_pedidos_ya, marker='o', label='Pedidos Ya')
plt.plot(years, user_base_hugo, marker='o', label='Hugo')
plt.plot(years, user_base_uber_eats, marker='o', label='Uber Eats')

plt.title('User Base Growth for Online Food Delivery Apps (2018-2023)')
plt.xlabel('Year')
plt.ylabel('User Base (thousands)')
plt.legend()
plt.grid(True)
plt.show()
```
