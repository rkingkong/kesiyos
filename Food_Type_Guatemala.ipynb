```python
import matplotlib.pyplot as plt

# Sample data
cuisines = ['Local', 'Italian', 'Asian', 'Fast Food', 'Healthy']
preferences = [35, 25, 15, 20, 5]  # percentage of consumers preferring each cuisine

plt.figure(figsize=(10, 6))
plt.bar(cuisines, preferences, color='skyblue')
plt.title('Consumer Preferences for Types of Cuisine in Guatemala')
plt.xlabel('Cuisine Type')
plt.ylabel('Percentage of Consumers')
plt.show()
```
